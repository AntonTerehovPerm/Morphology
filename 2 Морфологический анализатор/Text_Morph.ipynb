```python
def bigramm_chast_rechi(data_out):
    global word_rech
    word_rech = {}# посел, до #сохраняем биграммы частей речи
    #print(data_out)
    for x in data_out:
        text = parsing_from_slova_only_spisok_strok(x,['='])
        #print(text)
        last_rech = 'NI'
        for y in range(1,len(text),3):
            #print(text[y])
            if word_rech.get(last_rech)== None:
                word_rech[last_rech] = {'all':0}
                
            if word_rech.get(text[y][1])== None:
                word_rech[text[y][1]] = {'all':0} 
            
            if word_rech[last_rech].get(text[y][1]) == None:
                word_rech[last_rech][text[y][1]] = 1
            else:
                word_rech[last_rech][text[y][1]] += 1
            word_rech[last_rech]['all'] += 1
                
            if word_rech[text[y][1]].get(last_rech) == None:
                word_rech[text[y][1]][last_rech] = 1
            else:
                word_rech[text[y][1]][last_rech] += 1
            word_rech[text[y][1]]['all'] += 1
                
            last_rech = text[y][1]
```

```python
def clear_word(slovo):
    slovar_left_symbols = {',':0, '.':0, '?':0, '!':0}
    new_slovo = ''
    for x in slovo:
        if slovar_left_symbols.get(x) == None:
            new_slovo+=x
    return new_slovo
```

```python
def craft_slovar(out_text_slov):
    global slovar
    last = ' '
    slovar = {}
    for all_slov in out_text_slov:
        for one_slovo in all_slov:
            clear_slovo = clear_word(one_slovo).lower().strip()
            if slovar.get(clear_slovo) == None:
                    slovar[clear_slovo] = {'kol':1,'connection_word':{}, 'chast_rechi':{}, 'standart_form':''}
            else:
                    slovar[clear_slovo]['kol'] += 1
            if last == ' ':
                    last = clear_slovo
            else:
                if slovar[last]['connection_word'].get(clear_slovo) == None:
                    slovar[last]['connection_word'][clear_slovo] = 1
                else:
                    slovar[last]['connection_word'][clear_slovo] += 1
                last = clear_slovo
    #print(slovar)
```

```python
def percent_visible_word(out_text_slov):
    kol = 0
    length = 0
    for x in out_text_slov:
        length += len(x)
        for y in x:
            #print(y)
            if primer.get(y.lower()) != None:
                kol += 1
            #print(y.lower())
    print('Количество известных слов: ',kol,'\nДлина всего текста: ', length,'\nПроцент знания частей речи и форм из текста: ', kol/length*100)
    print('-'*100)
```

```python
def pars_sl(stroka):
    global open_core_slovar, chst, preob
    mass = stroka.split('>')
    out = [mass[0].split()[2][6:len(mass[0].split()[2])-1], mass[3].split()[2][3:len(mass[3].split()[2])-1],
           mass[4].split()[1][3:len(mass[4].split()[1])-2]]
    
    if chst.get(out[2]) == None:
        chst[out[2]] = 1
        
    if preob.get(out[2]) == None:
        out[2] = 'NI'
    else:
        out[2] = preob[out[2]]
        
    if open_core_slovar.get(out[0].lower()) == None:
        open_core_slovar[out[0].lower()] = out[2]
    
    if out[0].lower() != out[1]:     
        if open_core_slovar.get(out[1]) == None:
            open_core_slovar[out[1]] = out[2]
            
        
    return out
```

```python
def pars_corp(stroka):
    
    if (stroka.strip()[0:6] == '<token')and(stroka.strip() != '<tokens>'):#stroka.strip()[0:8] == '<source>' or 
        return pars_sl(stroka.strip())
        
```

```python
def checked_connections(slovar):
    kolich = 0
    for x in slovar:
        if slovar[x]['standart_form'] == '': kolich += 1
    return kolich
```

```python
def percent_chance(slo):
    '''
    Вычисляет вероятность для каждой части речи
    '''
    new_slo = {}
    #print('slo', slo)
    for s in slo:
        if s != 'all':
            new_slo[s] = slo[s]/slo['all']
    #print('new_slo',new_slo)
    return new_slo
```

```python
def percent_chance_l(slo):
    '''
    Вычисляет вероятность для каждой части речи
    '''
    new_slo = {}
    #print('slo', slo)
    for s in slo:
        if s != 'all':
            new_slo[s] = slo[s]
    #print('new_slo',new_slo)
    return new_slo
```

```python
def all_percent_word(mass_method):
    method_end = {}
    for num,x in enumerate(mass_method):
        if num == 0:
            method_end = x
        else:
            for y in x:
                if method_end.get(y) == None:
                    method_end[y] = x[y]
                else:
                    method_end[y] += x[y]
    print('method_end',method_end)
    return method_end
```

```python
def end_chast_rechi(mass_method):
    chast_rechi = [0,'']
    for x in mass_method:
        if chast_rechi[0] < mass_method[x]:
            chast_rechi[0] = mass_method[x]
            chast_rechi[1] = x
    print(chast_rechi[1])
    return chast_rechi[1]
```

```python
def vip_word(vip):
    global primer, spisok_chast_rechi_length, word_rech, slovar, word_end_chast_rechi, preob
    import pymorphy2
    morph = pymorphy2.MorphAnalyzer()
    new_vip = {}
    for x in vip:#предыдущие значения из вип
        z = 'NI'
        new_m = morph.parse(x)[0]
        if slovar.get(x) != None:
            if (len(slovar[x]['chast_rechi']) == 0)and(primer.get(x) != None):
                slovar[x]['chast_rechi'] = primer[x]['chast_rechi']
                slovar[x]['standart_form'] = primer[x]['standart_form']
                z = slovar[x]['chast_rechi']
                #print('z1',z)
                
            elif slovar[x]['standart_form'] == '':
                    #print('z2',z)
                    if len(slovar[x]['chast_rechi']) == 1:
                        slovar[x]['standart_form'] = x
                        z = slovar[x]['chast_rechi']
                        #print('z21',z)
                        
                    else:
                        chet = {}
                        for g in slovar[x]['chast_rechi']['bi']:
                            for u in g:
                                if chet.get(u) == None:
                                    chet[u] = g[u]
                                else:
                                    chet[u] += g[u]
                        #print('chet',chet)
                        
                        print(x)
                        print('chast_rechi',percent_chance(chet),'\n', 
                              'length',percent_chance_l(slovar[x]['chast_rechi']['l']),'\n',
                                                                   
                              'okonchania',slovar[x]['chast_rechi']['end'],'\n', 
                              'morphii',slovar[x]['chast_rechi']['morph'],'\n',
                                                                  
                              'corpora',slovar[x]['chast_rechi']['corp'],'\n')
                        
                        new_ch = end_chast_rechi(all_percent_word([percent_chance(chet), percent_chance_l(slovar[x]['chast_rechi']['l']),
                                                                   slovar[x]['chast_rechi']['end'], slovar[x]['chast_rechi']['morph'],
                                                                  slovar[x]['chast_rechi']['corp']]))
                        #print('new_ch', new_ch)
                        slovar[x]['standart_form'] = new_m.normal_form
                        slovar[x]['chast_rechi'] = new_ch
                        z = slovar[x]['chast_rechi']
                        #print('z22',z)
                        
           # print('z',z)           
            #print(x, slovar[x]['standart_form'])
            #print('MORPH')
            #print(x)
            #print(new_m.normal_form)
            #print(preob[str(new_m.tag.POS)])
            #print('+'*100)
            #print('ИТОГ')
            #print(slovar[x]['standart_form'])
            #print(slovar[x]['chast_rechi'])
            #print('-'*100)
            for y in slovar[x]['connection_word']:#список новых соединений
                if slovar[y]['standart_form'] == '':
                    if new_vip.get(y) == None:
                        new_vip[y] = 1
                        #print(slovar[x]['chast_rechi'])
                        #print(x, y)
                        if spisok_chast_rechi_length.get(len(y)) != None:
                            slovar[y]['chast_rechi']['l'] = spisok_chast_rechi_length[len(y)]
                        else:
                            slovar[y]['chast_rechi']['l'] = {'all':1, 'NI':1 }
                        
                        if (word_end_chast_rechi.get(y[len(y)-2:len(y)]) != None) and (len(y) > 2):
                            slovar[y]['chast_rechi']['end'] = {word_end_chast_rechi[y[len(y)-2:len(y)]]:1.5}
                        else:
                            slovar[y]['chast_rechi']['end'] = {'NI':0}
                        
                        if open_core_slovar.get(x) == None:
                            slovar[y]['chast_rechi']['corp'] = {'NI':0}
                        else:
                            slovar[y]['chast_rechi']['corp'] = {open_core_slovar[x]:1}
                        
                        new_y = morph.parse(y)[0]
                        slovar[y]['chast_rechi']['morph'] = {preob[str(new_y.tag.POS)]:3}
                        
                        #print(z)
                        if type(z) == type(dict()):
                            z = preob[str(new_m.tag.POS)]
                            slovar[x]['chast_rechi'] = z
                            
                        slovar[y]['chast_rechi']['bi'] = [word_rech[z]]
                    else:
                        new_vip[y] += 1
                        slovar[y]['chast_rechi']['bi'].append(word_rech[z])
    return new_vip
```

```python
def connection_check_out(slovar):
    vip = {}
    for x in slovar:#список новых соединений
                if slovar[x]['standart_form'] == '':
                    if vip.get(x) == None:
                        vip[x] = 1
                        slovar[x]['chast_rechi']['bi'] = {'NI':0}
    return vip
```

```python
def opred_morph():
    global primer, spisok_chast_rechi_length, word_rech, slovar
    check = checked_connections(slovar)
    vip = {}
    for x in primer:
        vip[x] = 1
    #print(vip)
    
    ls_check = checked_connections(slovar)
    while check >0:
        #print(check)
        #vip_word(vip)
        vip = vip_word(vip)
        #print(vip)
        #display(slovar)
        
        ls_check = check
        check = checked_connections(slovar)
        
        if ls_check == check:
            vip = vip_word(connection_check_out(slovar))
            check = checked_connections(slovar)
```
