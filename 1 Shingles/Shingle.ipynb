```python
import requests
import random as rd
import os
from tqdm import tqdm_notebook as tn
import time
import copy
```

```python
def create_url(N):
    '''
    Создает случайный номер длиной N
    '''
    user_id = 0
    for x in range(N):
        user_id += rd.randint(0,9)*(10**(x+1))
    user_id += rd.randint(0,9)
    return user_id
```

```python
def checked_404(direct,i):
    '''
    Проверяет страницу на ошибку
    '''
    f = open(direct+str(i)+'.txt','r',encoding='utf8')
    l = [line for line in f]
    k = 0
    for line in l:
        if line.strip() == '<h2 class="error__header">Запрошенная вами страница не существует или недоступна в данный момент.</h2>': k+=1
    if k>0: return False 
    else: return True
```

```python
def create_nice_symbols():
    mass = {}
    for x in range(1040,1104):
        #print(x,':',chr(x)) #1040 - 1103
        mass[chr(x)] = ''
    mass[' '] = ''    
    return mass
```

```python
def clear(text_r):
    global shingle
    mass_symbols = create_nice_symbols()
    clear_text = ''
    shingle_mass = []
    all_shingle = []
    #print(text_r)
    for x in text_r:
        #print('x',x)
        if mass_symbols.get(x) != None:

            if x == ' ' and clear_text.strip() != '' and clear_text.strip() != ' ':
                #print(clear_text)
                shingle_mass.append(clear_text)
                clear_text = ''
            else:
                clear_text += x.lower().strip()
                
            if len(shingle_mass) == shingle:
                all_shingle.append('_'.join(shingle_mass))
                shingle_mass = []
            
            #print(x)
            #print(clear_text)
    return all_shingle
```

```python
def parsing(j, error):
    direct = 'Data_News/'
    
    nomber = create_url(rd.randint(1,6))
    url = 'https://www.kommersant.ru/doc/'+str(nomber)
    r = str(nomber)+requests.get(url).text
    
    f = open(direct+str(j)+'.txt','w',encoding='utf8')
    f.write(r)
    f.close()
    
    if checked_404(direct,j): pass
    else: error += 1
    return error
```

```python
# Загрузка новостей
error = 0
call_states = 10000 #10000
#print(create_nice_symbols())
for j in tn(range(call_states)):
    error = parsing(j, error)
    time.sleep(1)
```

```python
def read_file_news(max_file_nomber):
        direct = 'Data_News/'
    
        for one in tn(range(1,max_file_nomber)):
            if  os.path.exists(direct+str(one)+'.txt'):
                
                r = open(direct+str(one)+'.txt','r',encoding='utf8')
                data_r = [line for line in r]
                data_r = [' '.join(data_r)]
                #print(data_r)
                #print(clear(data_r[0]))
                text_reader = clear(data_r[0])
                #print(text_reader)
                r.close()
                #print(text_reader)

                nomber = data_r[0].split('/n')[0]
                
                f = open('News/'+str(one)+'.txt','w',encoding='utf8')
                f.write(str(nomber)+'\n')
                for x in text_reader:
                    f.write(str(x)+'\n')

                    if os.path.exists('News/'+str(x)+'.txt'):
                        f_key = open('News/'+str(x)+'.txt','a',encoding='utf8')
                        f_key.write(str(one)+'\n')
                        f_key.close()
                    else:
                        try:
                            f_key = open('News/'+str(x)+'.txt','w',encoding='utf8')
                            f_key.write(str(one)+'\n')
                            f_key.close()
                        except:
                            pass

                f.close()
    
```

```python
def read_news(j,error):
    '''
    Парсит новости
    '''
    direct = 'News/'
    
    nomber = create_url(rd.randint(1,6))
    url = 'https://www.kommersant.ru/doc/'+str(nomber)
    r = str(nomber)+requests.get(url).text
    
    text_reader = clear(r)
    #print(r)
    
    f = open(direct+str(j)+'.txt','w',encoding='utf8')
    f.write(str(nomber)+'\n')
    for x in text_reader:
        f.write(str(x)+'\n')
        
        if os.path.exists(direct+str(x)+'.txt'):
            f_key = open(direct+str(x)+'.txt','a',encoding='utf8')
            f_key.write(str(j)+'\n')
            f_key.close()
        else:
            try:
                f_key = open(direct+str(x)+'.txt','w',encoding='utf8')
                f_key.write(str(j)+'\n')
                f_key.close()
            except:
                pass
            
    f.close()
    
    if checked_404(direct, j): pass
    else: error += 1
    return error
```

```python
call_states = 112
matrix = [[[0, False] for x in range(call_states)] for x in range(call_states)]
#matrix
```

```python
def craft_dict(data):
    new_slovar = {}
    for x in data:
        new_slovar[x] = ''
    return new_slovar
```

```python
def checked_bad_text(data_file_1, nomber_file_2):
    kol = 0
    all_len = 0
    file_2 = open('News/'+str(nomber_file_2)+'.txt','r',encoding='utf8')
    data_file_1 = craft_dict(data_file_1)
    for line in file_2:
        all_len += 1
        if data_file_1.get(line) != None:
            kol += 1
    return 100*kol/all_len
```

```python
def max_zhach(matrix):
    for nomber_i,x in enumerate(matrix):
        mas_max = [0,0]
        for nomber_j,y in enumerate(x):
            if nomber_i != nomber_j and mas_max[0] < y[0]:
                mas_max[0] = y[0]
                mas_max[1] = nomber_j
        print('Для ', nomber_i, 'наиболее совподащий файл на ', mas_max[0] ,'%  - это', mas_max[1] )
```

```python
for x in tn(range(call_states)):
    file = open('News/'+str(x)+'.txt','r',encoding='utf8')
    data = [line for line in file]
    file.close()
    
    for y in range(1,len(data)):
        try:
            file = open('News/'+str(data[y]).strip()+'.txt','r',encoding='utf8')
            for line in file:
                #print(line)
                if matrix[x][int(line)][1] == False:
                    matrix[x][int(line)][1] = True
                    matrix[x][int(line)][0] = checked_bad_text(data, int(line))
                    #убрать нижние строки, если и обратное проверить
                    matrix[int(line)][x][1] = True
                    matrix[int(line)][x][0] = checked_bad_text(data, int(line))
            file.close()
        except:
            pass
    
```
